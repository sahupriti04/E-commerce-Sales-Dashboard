```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Ecommerce_Sales_Data_2024_2025.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)

df.fillna("Unknown", inplace=True)

print("Missing Values After Cleaning:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(8,5))

sns.barplot(x='Category',
            y='Sales',
            data=df)

plt.title("Sales by Category")
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(x='Region',
            y='Profit',
            data=df)

plt.title("Profit by Region")

plt.show()
```

```python
plt.figure(figsize=(6,6))

df['Payment Mode'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Payment Mode Distribution")
plt.ylabel("")

plt.show()
```

```python
top_cities = df.groupby('City')['Sales'].sum() \
               .sort_values(ascending=False) \
               .head(5)

plt.figure(figsize=(8,5))

top_cities.plot(kind='bar')

plt.title("Top 5 Cities by Sales")
plt.xlabel("City")
plt.ylabel("Sales")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df['Sales'])

plt.title("Outlier Detection - Sales")

plt.show()
```

Key Insights

- Electronics and Clothing categories generated high sales.
- Certain regions produced better profits than others.
- Online payment methods were widely used.
- Some sales outliers were identified using boxplots.
